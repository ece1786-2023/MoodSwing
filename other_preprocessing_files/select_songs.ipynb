```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# FInd all files in folders relaxed, happy, sad, angry
import os
import glob
happy_files = glob.glob('happy/*.txt')
sad_files = glob.glob('sad/*.txt')
angry_files = glob.glob('angry/*.txt')
relaxed_files = glob.glob('relaxed/*.txt')
```

```python
# Import ml_balanced.xlsx
df = pd.read_excel('ml_balanced.xlsx')
df.head()
```

```python
# Create a dataframe for each emotion with each column 'Index', 'Artist', 'Title', 'Mood'
happy_df = pd.DataFrame(columns=['Index', 'Artist', 'Title', 'Mood'])
sad_df = pd.DataFrame(columns=['Index', 'Artist', 'Title', 'Mood'])
angry_df = pd.DataFrame(columns=['Index', 'Artist', 'Title', 'Mood'])
relaxed_df = pd.DataFrame(columns=['Index', 'Artist', 'Title', 'Mood'])
```

```python
# Get a list of all the indxes of the songs which is obtained from _files with after / before _
happy_index = []
for i in happy_files:
    happy_index.append(i.split('/')[1].split('_')[0])
sad_index = []
for i in sad_files:
    sad_index.append(i.split('/')[1].split('_')[0])
angry_index = []
for i in angry_files:
    angry_index.append(i.split('/')[1].split('_')[0])
relaxed_index = []
for i in relaxed_files:
    relaxed_index.append(i.split('/')[1].split('_')[0])
# Print firdt 5 indexes
print(happy_index[:5])
```

```python
# Get the artist and title of the songs and mood from the dataframe df and append it to the respective dataframes
happy_to_concat = []
# Your loop to append rows
for i in happy_index:
    happy_to_concat.append(df.loc[df['Index'] == i])
# Concatenate DataFrames outside the loop
happy_df = pd.concat(happy_to_concat, ignore_index=True)
# Sort the dataframe by index
happy_df.sort_values(by=['Index'], inplace=True)

sad_to_concat = []
# Your loop to append rows
for i in sad_index:
    sad_to_concat.append(df.loc[df['Index'] == i])
# Concatenate DataFrames outside the loop
sad_df = pd.concat(sad_to_concat, ignore_index=True)
# Sort the dataframe by index
sad_df.sort_values(by=['Index'], inplace=True)

angry_to_concat = []
# Your loop to append rows
for i in angry_index:
    angry_to_concat.append(df.loc[df['Index'] == i])
# Concatenate DataFrames outside the loop
angry_df = pd.concat(angry_to_concat, ignore_index=True)
# Sort the dataframe by index
angry_df.sort_values(by=['Index'], inplace=True)

relaxed_to_concat = []
# Your loop to append rows
for i in relaxed_index:
    relaxed_to_concat.append(df.loc[df['Index'] == i])
# Concatenate DataFrames outside the loop
relaxed_df = pd.concat(relaxed_to_concat, ignore_index=True)
# Sort the dataframe by index
relaxed_df.sort_values(by=['Index'], inplace=True)
```

```python
# Select the first 50 songs from each emotion and create a new dataframe
h_df = happy_df[:50]
s_df = sad_df[:50]
a_df = angry_df[:50]
r_df = relaxed_df[:50]
# Concatenate all the dataframes
frames = [h_df, s_df, a_df, r_df]
result = pd.concat(frames)
# Reset the index of the dataframe
result.reset_index(inplace=True)
# Drop the old index column
result.drop('index', axis=1, inplace=True)
# Save the dataframe as a csv file
# result.to_csv('selected_songs.csv', index=False)
result
```

```python
# Reset the index of the dataframe
h_df = h_df.reset_index(drop=True)
s_df = s_df.reset_index(drop=True)
a_df = a_df.reset_index(drop=True)
r_df = r_df.reset_index(drop=True)
```

```python
# Create a new folder named 'selected_lyrics' and create 4 subfolders named 'happy', 'sad', 'angry', 'relaxed' and save the lyrics of the songs in result in the respective folders
import os
import shutil
# Create a new folder named 'selected_lyrics'
# os.mkdir('selected_lyrics')
# # Create 4 subfolders named 'happy', 'sad', 'angry', 'relaxed'
# os.mkdir('selected_lyrics/happy')
# os.mkdir('selected_lyrics/sad')
# os.mkdir('selected_lyrics/angry')
# os.mkdir('selected_lyrics/relaxed')
# Copy the lyrics of the songs in result in the respective folders

# Happy songs that has i+_+ something before .txt
for i in range(len(h_df)):
    # print(h_df['Index'][i])
    location = glob.glob(f'happy/'+h_df['Index'][i]+'*.txt')
    shutil.copy(location[0], 'selected_lyrics/happy')
# Sad songs that has i+_+ something before .txt
for i in range(len(s_df)):
    location = glob.glob(f'sad/'+s_df['Index'][i]+'*.txt')
    shutil.copy(location[0], 'selected_lyrics/sad')
# Angry songs that has i+_+ something before .txt
for i in range(len(a_df)):
    location = glob.glob(f'angry/'+a_df['Index'][i]+'*.txt')
    shutil.copy(location[0], 'selected_lyrics/angry')
# Relaxed songs that has i+_+ something before .txt
for i in range(len(r_df)):
    location = glob.glob(f'relaxed/'+r_df['Index'][i]+'*.txt')
    shutil.copy(location[0], 'selected_lyrics/relaxed')
```
