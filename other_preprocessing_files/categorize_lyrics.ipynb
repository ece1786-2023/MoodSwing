```python
# The songDict dictionary has songIDs as keys and file names as values, 
# The df has a Index which matches the songIDs and a Mood column which has the mood of the song
# We will use this information to copy the files from the lyrics folder to the respective mood folders

# Loop through the songDict dictionary
for songID in songDict.keys():
    # Get the mood of the song from the df
    print(songID, df.index, sep=' | ', end='\n')
    # Find row in df with songID as index
    mood = df[df.Index == songID]['Mood'].values[0]
    # Get the file name from the songDict dictionary
    file = songDict[songID]
    # Get the file path
    filePath = 'lyrics/' + file
    print(filePath, mood, sep=' | ', end='\n')
    # Copy the file to the respective mood folder
    shutil.copy(filePath, mood)
```
