```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot graphs
def plot_graphs(df):
    # Plot train loss
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1)
    plt.plot(df['step'], df['train_loss'], label='Train Loss')
    plt.title('Train Loss')
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.legend()

    # Plot train accuracy
    plt.subplot(1, 3, 2)
    plt.plot(df['step'], df['train_accuracy'], label='Train Accuracy')
    plt.title('Train Accuracy')
    plt.xlabel('Steps')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot validation loss
    plt.subplot(1, 3, 3)
    plt.plot(df['step'], df['valid_loss'], label='Validation Loss')
    plt.title('Validation Loss')
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Read CSV file
csv_file_path = 'step_metrics.csv'  # Replace with your file path
df = pd.read_csv(csv_file_path)

selected_steps = range(1, df['step'].max() + 1, 10)
df_filtered = df[df['step'].isin(selected_steps)]

# Plot graphs
plot_graphs(df_filtered)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
csv_file_path = 'step_metrics.csv'  # Replace with your file path
df = pd.read_csv(csv_file_path)

# Filter data for steps 1, 11, 21, 31, etc.
selected_steps = range(1, df['step'].max() + 1, 10)
df_filtered = df[df['step'].isin(selected_steps)]

# Plot train_loss, valid_loss, and train_accuracy
plt.figure(figsize=(12, 5))

# Plot train_loss
plt.subplot(1, 2, 1)
plt.plot(df_filtered['step'], df_filtered['train_loss'], label='Train Loss', marker='o')
plt.title('Train Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot valid_loss
plt.subplot(1, 2, 2)
plt.plot(df_filtered['step'], df_filtered['valid_loss'], label='Validation Loss', marker='o')
plt.title('Validation Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot train_accuracy
plt.figure(figsize=(8, 5))
plt.plot(df_filtered['step'], df_filtered['train_accuracy'], label='Train Accuracy', marker='o', color='green')
plt.title('Train Accuracy')
plt.xlabel('Steps')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
